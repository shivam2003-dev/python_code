```python
import numpy as np


# create a 1d array
arr1 = np.array([1,2,3,4,5])
print(arr1)
print(type(arr1))
print(arr1.dtype)
print(arr1.ndim)
print(arr1.shape)
```

```python
# reshape into 2d
arr2 = np.array([1,2,3,4,5])
arr2 = arr2.reshape(1,5)               # 1 row 5 columns
print(arr2)
print(arr2.ndim)
print(arr2.shape)   
```
